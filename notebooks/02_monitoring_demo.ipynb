# 02 — Monitoring Demo: CoT, ToT, and Aha Moments

This notebook demonstrates the three monitoring components:
- **CoTMonitor** — Chain-of-Thought pattern detection
- **ToTMonitor** — Tree-of-Thought branch detection
- **AhaMonitor** — Aha-moment / breakthrough signal detection

```python
import sys
sys.path.insert(0, '..')

from src.monitoring.cot_monitor import CoTMonitor
from src.monitoring.tot_monitor import ToTMonitor
from src.monitoring.aha_monitor import AhaMonitor
import json
```

```python
sample_outputs = [
    "Step 1: Identify the given information. Step 2: Apply the formula. Step 3: Compute. Therefore: The answer is 42.",
    "Option 1: Greedy — fast but may miss optimal. Option 2: DP — optimal but slower. Best option: Option 2.",
    "I was computing this wrong. Wait, actually — I see now that the constraint changes everything.",
    "The answer is 5.",
]
```

## Chain-of-Thought Monitor

```python
cot = CoTMonitor(log_every=0)
trace = cot.extract(sample_outputs[0])
print(f"Has CoT: {trace.has_cot}  |  Steps: {trace.step_count}")
for step in trace.steps:
    print(f"  [{step.index}] conclusion={step.is_conclusion}: {step.text[:60]}")

cot.update(sample_outputs)
print("\nStats:", json.dumps(cot.get_stats(), indent=2))
```

## Tree-of-Thought Monitor

```python
tot = ToTMonitor(log_every=0)
trace = tot.extract(sample_outputs[1])
print(f"Has ToT: {trace.has_tot}  |  Branches: {trace.branch_count}")
for node in trace.nodes:
    status = "SELECTED" if node.is_selected else "PRUNED" if node.is_pruned else "-"
    print(f"  [{node.node_id}] {status}: {node.text[:60]}")

tot.update(sample_outputs)
print("\nStats:", json.dumps(tot.get_stats(), indent=2))
```

## Aha-Moment Monitor

```python
aha = AhaMonitor(log_every=0)
trace = aha.extract(sample_outputs[2])
print(f"Has Aha: {trace.has_aha}  |  Moments: {trace.moment_count}")
for moment in trace.moments:
    print(f"  [{moment.index}] [{moment.trigger_pattern}] pos={moment.position}")

aha.update(sample_outputs)
print("\nStats:", json.dumps(aha.get_stats(), indent=2))
```
